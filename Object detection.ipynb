```python
from PIL import Image
import io
import matplotlib.pyplot as plt

# Get the binary data of the uploaded image
image_data = next(iter(uploaded.values()))

# Open the image using PIL
image = Image.open(io.BytesIO(image_data))

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```

```python
import matplotlib.patches as patches

# Example bounding boxes and labels for object detection
bounding_boxes = [(50, 50, 200, 200), (250, 250, 400, 400)]  # Example coordinates (x1, y1, x2, y2)
labels = ["cat (95%)", "dog (90%)"]  # Example labels with confidence scores

fig, ax = plt.subplots(1)
ax.imshow(image)

for bbox, label in zip(bounding_boxes, labels):
    x1, y1, x2, y2 = bbox
    rect = patches.Rectangle((x1, y1), x2-x1, y2-y1, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    plt.text(x1, y1, label, color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))

plt.axis('off')
plt.show()
```

Below are the explanation of the steps involved in uploading, reading, and displaying an image in Google Colab.

Step-by-Step Guide:
Uploading the Image:

First, i imported the files module from google.colab to use the file upload functionality.
i used the files.upload() method to upload a file. This opened a file dialog in the Colab interface where i selected the image to upload.
The uploaded files stored in a dictionary where the keys are the filenames and the values are the file content in bytes.

Reading the Image:

After uploading, i extracted the image data from the uploaded files.
i Used the next(iter(uploaded.values())) to get the first (and usually the only) uploaded file content.
i opened the image using the PIL (Python Imaging Library) which is now part of the Pillow package. This involves converting the byte data to an image object.

Displaying the Image:

To display the image, i used matplotlib library. This library provided a pyplot module for plotting and visualizing data, including images.

Use plt.imshow(image) to display the image.
Use plt.axis('off') to hide the axis for a cleaner look.
Finally, call plt.show() to render the image.

This complete workflow allowed me to upload an image, read it, and display it in Google Colab, making it easier to visualize and further process the image data.
